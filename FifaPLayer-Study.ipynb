```python
#Histogram showing player ages shows typicall age for players is around 23
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('FifaData.csv')
age= df.Age
list(df)
plt.hist(age, density=True, bins=25, color="green")
plt.title('Fifa Player Ages')

plt.xlabel('Ages')
plt.show()
```

```python
#Scatter plot to show 10 countries player wages
plt.scatter(
    x=df.Wage.head(10),
    y=df.Nationality.head(10),
    color="purple",
    s=60,
)
plt.title('Fifa Player Wages')
plt.xlabel('Player Wages')
plt.ylabel("Country")
plt.show()
```

```python
#Boxplot using ages shows the age ranges for the players avg is in the 20's and few players ages are >=40
plt.boxplot(age)
plt.ylabel("Age")
plt.title("Fifa Player Age")
plt.show()
```
