## DOB ECB Fines- a Research Proposal
The Department of Buildings (DOB) in New York City regulates City Construction Codes, Zoning, and  Dwelling Laws for over one million construction sites and buildings in New York City. Their aim is to enforce compliance to promote safety for site workers and the public.

With annual reviews and site inspections Violations are issued across the city for a range of different infractions.

According to data dating back to 1910 from the New York City Open Data source these DOB ECB (Enviornmental Control Board) violations have increased significantly in the last 25 years. 

Not only has there been an increase in the number of violations respondents receive annually, but there has been a significant change to the initial imposed penalty amounts. Now if we factor in inflation we may find some significance there as well. The biggest change is the number of violations and their respective outcomes. What happens after the violation is issued?

A violation is issued to a ‘Respondent’; a person or entity that the City names as responsible for the summons and is responsible for answering the summons. In most cases the Respondent is named as what the NYC Department of Finance recognizes as ownership of the place of occurrence. However, a ’Respondent’ is not necessarily located in the Building receiving the violation. In fact several Owner’s, Property Manager’s, Landlord’s and Respondent’s physical addresses are out of state. 


After analyzing the data from the past 25-30 years we can see that most Violations that where the Respondent either cured the violation before the hearing date or attended the scheduled court appearance result in a $0 penalty. In some cases, however, the violation will result in a **negative balance** for the respondent. 

When there is a negative balance we know there is an error in the system. This could tell us that the Respondent overpaid a Violation’s balance, but this would also be caused by a communication gap. The main take away here is that a negative balance essentially means that the Respondent is owed money from the city either way. 

Negative balances and missed court appearances are one reason any violation with a cure, compliance or scheduled hearing date should be received by the Respondent in a timely fashion. When it is recieved and acknowledged by the Respondent, Property Owner, Landlord or Tenant it gives the Respondent time to address the violation and meet or show compliance while keeping the costs at a minimum for both the city and the Respondent.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_rows = 3000
%matplotlib inline
```

```python
dob = pd.read_csv('DOB_ECB_Violations.csv')
dob.shape
```

```python
#From dataset "Borough1= Manhattan, 2= Bronx, 3 = Brooklyn, 4 = Queens, 5 = Staten Island"
dob['BORO'] = dob['BORO'].replace(1,'Manhattan').replace(2,'Bronx').replace(3,'Brooklyn').replace(4,'Queens').replace(5,'Staten Island').replace(6,'unknown').replace(3012920, 'unknown')
```

```python
#Accounted for one value that was entered as '21'
dob = dob.loc[dob['HEARING_DATE']>21]
```

```python
#clean up date format for hearing date and issue date
dob['new_hearing_date']=pd.to_datetime(dob['HEARING_DATE'], format='%Y%m%d');
```

```python
dob = dob.loc[dob['ISSUE_DATE'] > 0]
```

```python
dob['new_issue_date'] = pd.to_datetime(dob['ISSUE_DATE'], format='%Y%m%d');
```

```python
#look for null values
dob[dob['HEARING_DATE'].isnull()]
```

```python
#remove missing values (NaN)
dob.loc[dob['PENALITY_IMPOSED'].isna(),'PENALITY_IMPOSED'] = 0
dob.loc[dob['AMOUNT_PAID'].isna(),'AMOUNT_PAID'] = 0
dob.loc[dob['BALANCE_DUE'].isna(),'BALANCE_DUE'] = 0
dob.loc[dob['BIN'].isna(),'BIN'] = 'Unknown'
dob.loc[dob['RESPONDENT_NAME'].isna(),'RESPONDENT_NAME'] = 'Unknown'
dob.loc[dob['RESPONDENT_STREET'].isna(),'RESPONDENT_STREET'] = 'Unknown'
dob.loc[dob['RESPONDENT_CITY'].isna(),'RESPONDENT_CITY'] = 'Unknown'
dob.loc[dob['RESPONDENT_ZIP'].isna(),'RESPONDENT_ZIP'] = 'Unknown'
dob.loc[dob['RESPONDENT_HOUSE_NUMBER'].isna(),'RESPONDENT_HOUSE_NUMBER'] = 'Unknown'
```

After analyzing the data from the past 25-30 years we can see that most Violations that have either been cured before the hearing date or attended the scheduled court date result in a $0 penalty. In some cases, the violations have resulted in a **negative balance** for the respondent. 

This could mean that the Respondent overpaid on a Violation’s balance, but essentially it means that the Respondent is owed the negative balance from the city. In any case, a Violation’s balance will change overtime if not addressed by the hearing date. This would cause the violation to accrue additional fees or interest over time. 

This is why it any violation with a cure, compliance or scheduled hearing date should be received by the Respondent in a timely fashion. It will give them time to meet or show compliance, and keep the costs at a minimum for both the city and the Respondent.
